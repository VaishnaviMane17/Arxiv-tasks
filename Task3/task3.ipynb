# Project2_Task3

##### Importing dataframe from task1 csv file.  

```python
import pandas as pd
import numpy as np
pd.options.mode.chained_assignment = None
df = pd.read_csv ('Metadata.csv')
```

##### Seperating dataframe for papers with required primary category.

##### Added a new column named no_of_authors derived by length of list in column author

```python
df_LG=df.loc[df['primary_category'] =='cs.LG']
df_LG.loc[:,'no_of_author']= df_LG['authors'].str.split(",").str.len()
```

##### Created a column named author type with the help of numpy, using condition list and value list

```python
conditions=[(df_LG['no_of_author']==1),(df_LG['no_of_author']==2),(df_LG['no_of_author']==3)|(df_LG['no_of_author']==4),df_LG['no_of_author']>4]
values = ['single author papers', 'double author papers', '3-4 authors papers', 'More than four authors']
df_LG['autor_type'] = np.select(conditions, values)
df_LG.head()

```

### Pie chart for cs.LG stating percentage of author types.

```python
df_LG.groupby(['autor_type']).sum().plot(kind='pie', autopct='%1.0f%%',
                                colors = ['red', 'green', 'yellow','blue'],legend=False,subplots=True)
```
