```python
import arxiv
import numpy as np
import itertools
import pandas as pd
```

### Our task required extraction of data from arxiv website.

#### It can be done with the help of arxiv Ids- A no. created with year and month of publication and Number of paper for that month Starting from 00001 which can go upto 99999.
#### I have created the Id_list with the help of numpy for every year seperately.Further, extracted metadata for earch IDs with the help of the id_list.


# Extracting Data for Year 2017

```python
id_list17=[]
for i in np.arange(170100000,171300000,1):
    j=np.round(i*10**-5,5)
    k=format(j,'.5f')    #adds specific no. of zeros after decimal (output type string)
    id_list17.append(k)

len(id_list17)
```

#### If I try to extract metadata for the entire length of id_list, it throws an http 414 error.With trial and error, I came to an conclusion that the maximum limit was 5oo Ids at a time.Hence breaking 2017 IDs into batch of 500 and looping for the whole length of id_list. 


```python
low = 0
high = 500
break_flag = False
ctr = 0
Metadata2017=[]
while low != len(id_list17):
    search = arxiv.Search(id_list=id_list17[low:high])
    result= list(search.results())
    for i in range(len(result)):
        paper_info={}
        if result[i].primary_category in ['cs.LG','cs.AI','cs.CC','cs.AR']:   # Condition to extract only for required primary category  
            #Title of paper
            paper_info['title'] = result[i].title
            print("title:", paper_info['title'])

            #authors (they have to be taken one by one and converted into string)
            paper_info['authors'] = []

            for author in result[i].authors:
                paper_info['authors'].append(str(author))

            print("authors:",paper_info['authors'])

            #summary (abstract)
            paper_info['summary'] = result[i].summary
            print("summary/abstract:",paper_info['summary'])

            #comment by author
            paper_info['comment'] = result[i].comment
            print("comment:",paper_info['comment'])

            #journal reference (if present)
            paper_info['journal_ref'] = result[i].journal_ref
            print("journal_ref:",paper_info['journal_ref'])

            #DOI (if present)
            paper_info['doi'] = result[i].doi
            print("doi:",paper_info['doi'])
            
            #entry_id
            paper_info['entry_id'] = result[i].entry_id
            print("entry_id:", paper_info['entry_id'])

            #last upated
            paper_info['updated'] = str(result[i].updated)
            print("last updated:",paper_info['updated'])

            #first published date
            paper_info['published'] = str(result[i].published)
            print("original posting date:",paper_info['published'])


            #Primary category
            paper_info['primary_category'] = result[i].primary_category
            print("primary category:", paper_info['primary_category'])

            #All categories
            paper_info['categories'] = result[i].categories
            print("categories:", paper_info['categories'])

            #links
            paper_info['links'] =  str(result[i].links)
            print("links:",paper_info['links'])

            #pdf_url
            paper_info['pdf_url'] = result[i].pdf_url
            print("pdf_url:",paper_info['pdf_url'])
            if paper_info!={}:
                Metadata2017.append(paper_info)
            else:
                break
    if result == []:
        break_flag = True #Set flag that loop was broken out of
        break
    ctr_by_type17 = {'cs.LG':0,'cs.AI':0,'cs.CC':0,'cs.AR':0}
    for i in range(len(result)):
    #If result in desired category
        if result[i].primary_category in ['cs.LG','cs.AI','cs.CC','cs.AR']:
            ctr += 1 #Increment counter
            ctr_by_type17[result[i].primary_category] += 1 
        print("Range:",(low,high))
        print("Cumulative count:",ctr)
        print("Count by type",ctr_by_type17)
        print("-----")
        low = high  #Old high becomes new low
        high += 500 #New high incremented by 500
    
    
```
